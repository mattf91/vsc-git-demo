```python
def greet(name, language):
    if language == "it":
        return f"Ciao, {name}!"
    elif language == "en":
        return f"Hello, {name}!"
    elif language == "es":
        return f"Hola, {name}!"
    else:
        return f"Hi, {name}!"

if __name__ == "__main__":
    print(greet("Matteo", "it"))
```
